```python
import pandas as pd
from datetime import datetime
import seaborn as sns 
import matplotlib.pyplot as plt
from pandas.plotting import table

import numpy as np

#file with the information from the process RE - SIE from 2017 to 2020 
dfcontratos = pd.read_csv("contratos-unificados-contipocontribuyente.csv",converters={'Ruc': lambda x: str(x)})
```

```python

x = range(2017, 2021, 1)
dfunified = pd.DataFrame()

#
for year in x: 
    newdf2 = dfcontratos[(dfcontratos.Year == year)]
    
    newdf2 = newdf2.groupby(['Company','Year','Ruc','Type']).agg({'TotalCost':['sum','count']})
    newdf2.sort_values(by=[('TotalCost','sum')], inplace=True)

    df = newdf2[newdf2.shape[0]-10:]
    
    df=df.reset_index()

    df.columns = [
      'Company',
      'Year',
      'Ruc',
      'Type',
      'TotalCostSum',
      'NumContratos'
    ]
    print(year)

    fig, ax = plt.subplots()
    # the size of A4 paper
    fig.set_size_inches(9.7, 7.27)
    sns.barplot(x = 'TotalCostSum',
                y = 'Company',
                hue = 'Type',
                data = df,
               )

    plt.show()

#This Graph display the 10 companies/individuals with the biggest contracts per year
```

```python
df = dfcontratos.groupby(['Year', 'Type']).agg({'TotalCost':['sum','count']})
#df.sort_values(by=[('TotalCost','sum')], inplace=True)
df = df.reset_index()
df.columns = [
  'Year',
  'Type',
  'TotalCostSum',
  'Contracts'
]

df2 = df.pivot(index='Year', columns='Type', values='TotalCostSum')
df2['IncreasingPN']=0
df2['IncreasingS']=0


pnTemp = 0
sTemp = 0
for i, row in df2.iterrows():
   
    if i == 2017:
       pnTemp = row['PN']
       sTemp = row['S']
       df2.loc[i, 'IncreasingPN'] = 0
       df2.loc[i, 'IncreasingS'] = 0
    
    increasingPN = row['PN'] - pnTemp
    increasingS = row['S'] -sTemp
    
    
    df2.loc[i, 'IncreasingPN'] = increasingPN
    df2.loc[i, 'IncreasingS'] = increasingS
    
    pnTemp = row['PN']
    sTemp = row['S']
    

print(df2)
```

```python
df3 = df2.iloc[:, 2:4]
df3
```

```python
#df3 = df3.melt('Year', var_name='cols',  value_name='vals')
df3 = df3.reset_index()
print(df3)
df_unpivoted = df3.melt(id_vars='Year', var_name='Type', value_name='Increasing')
df_unpivoted.sort_values(by=['Year'], inplace=True)
df_unpivoted
```

```python
fig, ax = plt.subplots()
# the size of A4 paper
fig.set_size_inches(11.7, 8.27)
sns.lineplot(x="Year", y="Increasing",
             hue="Type",
             data=df_unpivoted).set_title('Incremento de gastos por ao desde 2017 hasta mediados 2020')

#This graph displays the increase of expenses between 2017 to 2020
```
